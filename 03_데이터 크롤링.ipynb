```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# 1.  text를 돔트리로꾼 다음 
# 2.selector : 데이터 추출(뉴스의 제목막)

#c114-panel > div > c-wiz > div > c-wiz:nth-child(1) > c-wiz > article > div.m5k28 > div.XlKvRb > a

res = requests.get('https://finance.naver.com/item/news.naver?code=328130')
html = res.text 
print(html) # 문자열 반환
```

```python
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
my_title=soup.find_all('a',attrs={'class':'tit'})
print(my_title)
```
